```python
# data preparation
import pandas as pd
import numpy as np
from numpy import quantile, where, random
# data visualzation
import matplotlib.pyplot as plt
import seaborn as sns
# outlier/anomaly detection
from sklearn.neighbors import LocalOutlierFactor
from sklearn.cluster import KMeans 
```

```python
# data
data = pd.read_csv("3.csv")
df = data[["sepal_length", "sepal_width"]]
df1=df.sample(frac=1).reset_index(drop=True)
k = len(df1) / 10
c1 = df1[0:54000]
c2 = df1[54001:108000]
c3 = df1[108001 : 162000]
c4 = df1[162001 : 216000]
c5 = df1[216001 : 270000]
c6 = df1[270001 : 324000]
c7 = df1[324001 : 378000]
c8 = df1[378001 : 432000]
c9 = df1[432001 : 486000]
c10 = df1[486001 : 540000]
ct = df1[0:540000]
```

```python
# plot data points

plt.scatter(c1["sepal_length"], c1["sepal_width"], color = "b", s = 5)
plt.grid()
```

```python
# model specification
model1 = LocalOutlierFactor(n_neighbors = 3, metric = "manhattan", contamination = 0.07)
# model fitting
y_pred = model1.fit_predict(c1)
# filter outlier index
outlier_index1 = where(y_pred == -1) # negative values are outliers and positives inliers
# filter outlier values
outlier_values1 = c1.iloc[outlier_index1]
# plot data
plt.scatter(c1["sepal_length"], c1["sepal_width"], color = "b", s = 5)
# plot outlier values
plt.scatter(outlier_values1["sepal_length"], outlier_values1["sepal_width"], color = "r")
plt.grid()
```
